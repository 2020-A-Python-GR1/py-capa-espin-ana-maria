```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
```

```python
path = "./datasuicidios.csv"
```

```python
lista = pd.read_csv(
        path
        )
```

#### 1. ¿Que datos utilizaremos para el análisis?

```python
lista = lista.sort_values('year')
lista = lista.reset_index(drop=True)
```

```python
p = sns.barplot(x='sex', y='suicides_no', data=lista)
```

#### 3. Gráfica de suicidios por rango de edad

#### 4. Suicidios por género y rango de edad

```python
p = sns.barplot(x='sex', y='suicides_no', hue='age', data=lista)
```

#### 5. De la data generar estadísitca por año y edad

```python
def count_year(group):
    return len(group.year.unique())
country_o = lista.groupby('country').apply(count_year).sort_values(ascending = False)
country_o.name = 'occurances of country'
nmaster = lista[~ (lista.year == 2016)].drop(['HDI for year','country-year','suicides/100k pop'],axis = 1)
coun_year = lista.groupby('country-year').population.sum()
ncountry_o = nmaster.groupby('country').apply(count_year).sort_values(ascending = False)
ncountry_o[country_o >= 30] 
pmaster = nmaster[nmaster.country.isin(ncountry_o[country_o >= 31].index)]
age_group = pd.pivot_table(pmaster,index=['year'],values='suicides_no',columns=['age'],aggfunc=np.sum)
sui_age=age_group[['5-14 years','15-24 years', '25-34 years', '35-54 years','55-74 years', '75+ years']]
sui_age.plot(kind = 'bar',figsize=(10,6),stacked=True,title='Número de Suicidos por año y por rando de edad')
```

#### 10. ¿Cuál es el numero total de victimas por país?

```python
suicide_sum = lista.groupby('country', sort=False)["suicides_no"].sum().reset_index(name ='total')
suicide_sum = suicide_sum.sort_values(by="total", ascending=False)

suicide_sum.head()
```

#### 11.  Número total de victimas por rango de edad

```python
suicide_sum = lista.groupby('age', sort=False)["suicides_no"].sum().reset_index(name ='total')
suicide_sum = suicide_sum.sort_values(by="total", ascending=False)

suicide_sum.head()
```

#### 12. Número total de victimas por género

```python
suicide_sum = lista.groupby('sex', sort=False)["suicides_no"].sum().reset_index(name ='total')
suicide_sum = suicide_sum.sort_values(by="total", ascending=False)

suicide_sum.head()
```

#### 13. Número total de victimas por generación

```python
suicide_sum = lista.groupby('generation', sort=False)["suicides_no"].sum().reset_index(name ='total')
suicide_sum = suicide_sum.sort_values(by="total", ascending=False)

suicide_sum.head()
```

#### 14. Número total de victimas por año

```python
suicide_sum = lista.groupby('year', sort=False)["suicides_no"].sum().reset_index(name ='total')
suicide_sum = suicide_sum.sort_values(by="total", ascending=False)

suicide_sum.head()
```

#### 15.  Comparacion año versus victimas utilizando el campo gdp_percapita

```python
data_gdp = (lista['gdp_per_capita ($)'].groupby(lista.year)).sum()
data_gdp.plot(figsize=(20,10), linewidth=2, fontsize=15,color='green')
plt.xlabel('Year', fontsize=15)
plt.ylabel(' Total gdp_per_capita ($)',fontsize=15)
plt.show()
```

#### 16. Tendencia por edad vs gdp_per_capita

```python
data_gdp = (lista['gdp_per_capita ($)'].groupby(lista.age)).sum()
data_gdp.plot(figsize=(20,10), linewidth=2, fontsize=15,color='blue')
plt.xlabel('age', fontsize=15)
plt.ylabel(' Total gdp_per_capita ($)',fontsize=15)
plt.show()
```

#### 17. Tendencia por país

```python
data_gdp = (lista['gdp_per_capita ($)'].groupby(lista.country)).sum()
data_gdp.plot(figsize=(20,10), linewidth=2, fontsize=15,color='red')
plt.xlabel('pais', fontsize=15)
plt.ylabel(' Total gdp_per_capita ($)',fontsize=15)
plt.show()
```

#### 18. Grafica en pastel

```python
pv_primeros= lista.filter(items = ["country", "year"]).sort_values('year', ascending=False).head(20)

plt.figure(figsize=(16,8))
ax1 = plt.subplot(121, aspect='equal')

pv_primeros.plot(kind='pie', y = 'year', ax=ax1, autopct='%1.1f%%', 
 startangle=100, shadow=False, labels=pv_primeros['country'], fontsize=10)
plt.axis('equal')

plt.show()
```

#### 19. Comparación de suicidios por país

```python
data_suicide_mean = lista['suicides/100k pop'].groupby(lista.country).mean().sort_values(ascending=False)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(data_suicide_mean.head(10).index,data_suicide_mean.head(10),palette='coolwarm')
```

#### 20. Suicidios por año en gráfica pastel

```python
pv_primeros= lista.filter(items = ["year", "suicides_no"]).sort_values('suicides_no', ascending=False).head(20)

plt.figure(figsize=(16,8))

ax1 = plt.subplot(121, aspect='equal')

pv_primeros.plot(kind='pie', y = 'suicides_no', ax=ax1, autopct='%1.1f%%', 
 startangle=100, shadow=False, labels=pv_primeros['year'], fontsize=10)
plt.axis('equal')

plt.show()
```
