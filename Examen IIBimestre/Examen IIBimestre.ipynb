#### 7) Crear un Dataframe con 10 registros y 6 columnas y Ordenar el dataframe 1 vez por cada columna, ascendente y descendente

```python
df7 = pd.DataFrame(np.random.randint(0,10,60).reshape(10,6))
df7
```

```python
for i in range(0,6):
    ascendente = df7.sort_values(by=i, ascending=True)
    print('\n'+str(i)+'\n')
    print(ascendente)
```

```python
for i in range(0,6):
    descendente = df7.sort_values(by=i, ascending=False)
    print('\n'+str(i)+'\n')
    print(descendente)
```

#### 12) Crear un Dataframe con 10 registros y 6 columnas llenas de strings. Luego, unir la columna 1 y 2 en una sola, la 3 y 4, y la 5 y 6 concatenando su texto.

```python
strings = np.random.choice(['uno', 'dos', 'tres', 'cuatro', 'cinco', 'seis'], 60).reshape(10,6)
df12 = pd.DataFrame(strings)
df12
```

```python
a = df12[0].append(df12[1])
b = df12[2].append(df12[3])
c = df12[4].append(df12[5])
dffinal = pd.DataFrame(a)
dffinal[1] = b
dffinal[2] = c
dffinal
```
